Overview skim:
- Maximum entropy
  - Normal distribution

- Generalized linear models
  - Link functions, bounding probabilities
  - Exponential distributions, strictly positive
  - Gamma distributions
  - Poison distributions
  - Logit link
- Maximum entropy priors
- Conceptual chapter
  - "Information enetropy is essentially a measure of the number of ways a distribution can arise, according to state assumptions."

- E.g., knots. More ways to be tangled than untied
- "Bet on the distribution with the biggest entropy"
- Largest entropy = spreading probability as evenly as possible
- Generalized linear models
  - Like linear regressions, but replaces parameter of a likelihood function with a linear model
  - But no need to use Gaussian likelihoods
- "The distribution that can happen the most ways is also the distribution with the biggest information entropy. The distribution with the biggest entropy is the most conservative distribution that obeys its constraints."
- Example: 5 buckets, 10 individually numbered pebbles.
